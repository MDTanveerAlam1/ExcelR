#**1. State the Hypotheses:**

**Null Hypothesis (H0):** The theoretical weekly operating cost model accurately represents the operating costs.*(H<sub>0</sub>)* : Œº = 1000 + 5 ùñ∑ 600 = 4000


Alternative Hypothesis (H1): The weekly operating costs are higher than the model suggests.*(H<sub>1</sub>)* : Œº > 4000

#**2. Calculate the Test Statistic:**



*   Sample mean (x) = Rs. 3,050
*   Theorectical mean (Œº) = Rs. 4,000
*   Standard deviation (œÉ) = 5*25 = 125
*   Sample size(n) = 25

Formula for the testt statistics(t):

              t=(x-Œº)/(œÉ/‚àön)




```python
import math

# Given values
sample_mean = 3050
theoretical_mean = 4000
std_deviation = 125
sample_size = 25

# Calculate the test statistic
t = (sample_mean - theoretical_mean) / (std_deviation / math.sqrt(sample_size))
print(f"Test Statistic (t): {t}")
```

```python
import scipy.stats as stats

# Calculate the test statistic
t = (sample_mean - theoretical_mean) / (std_deviation / math.sqrt(sample_size))
```

#**3. Determine the Critical Value:**

Using an alpha level of 5% (Œ± = 0.05), we find the critical value from the standard normal (Z) distribution table. For a one-tailed test at Œ± = 0.05, the critical Z value is approximately 1.645.

```python
# Critical value for alpha = 0.05 (one-tailed test)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)
print(f"Critical Value (Z): {critical_value}")
```

#**4. Make a Decision:**

Compare the calculated test statistic with the critical value.
* If t > 1.645 , we reject the null hypothesis.
* Otherwise, we fail to reject the null hypothesis.

```python
# Make a decision
if t > critical_value:
    decision = "Reject the null hypothesis. There is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests."
else:
    decision = "Fail to reject the null hypothesis. There is not enough evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests."
```

#**5. Conclusion:**

```python
print(f"Test Statistic (t): {t}")
print(decision)
```

#**Graph the Results**

To visualize the hypothesis test, Create a plot showing the critical region and the test statistic:

```python
import matplotlib.pyplot as plt
import numpy as np

# Values for the plot
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

# Plot the normal distribution
plt.plot(x, y, label='Standard Normal Distribution')

# Shade the critical region
plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.5, label='Critical Region')

# Plot the test statistic
plt.axvline(t, color='blue', linestyle='--', label='Test Statistic (t)')

# Add labels and legend
plt.title('Hypothesis Testing: Critical Region and Test Statistic')
plt.xlabel('Test Statistic')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```
